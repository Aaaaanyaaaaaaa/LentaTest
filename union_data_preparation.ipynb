В этом файле мы объединим два датасета 'df_result_sums.csv' и 'df_result_chqs.csv'. Объединение будет straightforward. Мы возьмем среднее по всем колонкам, кроме gone.  Для колонок gone используем логическое И

```python
df_result_sums = pd.read_csv('df_result_sums.csv')
```

```python
df_result_sums.head()
```

```python
df_result_chqs = pd.read_csv('df_result_chqs.csv')
```

```python
df_result_chqs.head()
```

```python
common_clients= np.array(list(set(df_result_chqs.client_id.values) &  set(df_result_sums.client_id.values) ))
                                                       
```

```python
len(common_clients), len(df_result_chqs.client_id.values), len(df_result_sums.client_id.values)
```

```python
df_result_sums_cut = df_result_sums[df_result_sums.client_id.isin(common_clients)]
```

```python
df_result_chqs_cut = df_result_chqs[df_result_chqs.client_id.isin(common_clients)]
```

```python
df_result_chqs_cut_sorted = df_result_chqs_cut.sort_values(by=['client_id'])

df_result_sums_cut_sorted = df_result_sums_cut.sort_values(by=['client_id'])
```

```python
df_result_sums_cut_sorted
```

```python
df_result_chqs_cut_sorted
```

```python
False in (df_result_chqs_cut_sorted.client_id.values == df_result_sums_cut_sorted.client_id.values)
```

```python
df_result_chqs_cut_sorted.columns
```

```python
merged = pd.merge(df_result_chqs_cut_sorted[['total_num_chqs', 'total_num_days',
       'num_unique_goods', 'total_money_spent', 'total_is_promo_sum',
       'trend_coef', 'sparsity', 'gone']],
         df_result_chqs_cut_sorted[['total_num_chqs', 'total_num_days',
       'num_unique_goods', 'total_money_spent', 'total_is_promo_sum',
       'trend_coef', 'sparsity', 'gone']],
         on = df_result_sums_cut.client_id)
```

```python
merged
```

```python
resulting_df = pd.DataFrame()
resulting_df['client_id'] = merged['key_0']
```

```python
cols = ['total_num_chqs', 'total_num_days',
       'num_unique_goods', 'total_money_spent', 'total_is_promo_sum',
       'trend_coef', 'sparsity']
```

```python
for col_name in cols:
    
    col_name_x = col_name + '_x'
    col_name_y = col_name + '_y'
    
    resulting_df[col_name] = (merged[col_name_x] + merged[col_name_y])/2
```

```python

resulting_df['gone'] = merged['gone_x'] & merged['gone_y']
```

```python
resulting_df['gone'].sum()/resulting_df['gone'].count()
```

```python
resulting_df.to_csv('mean_data_of_sums_chqs.csv')
```
